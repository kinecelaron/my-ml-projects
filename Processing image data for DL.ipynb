```python
# getting an image using web get

import requests

url = 'https://tractive.com/blog/wp-content/uploads/2016/04/puppy-care-guide-for-new-parents.jpg'
response = requests.get(url)

# Save the content of the response as a file
with open('puppy-care-guide-for-new-parents.jpg', 'wb') as file:
    file.write(response.content)
```

```python
grayscale_image = cv2.imread('C:/Users/Dmwin/Classification & Regression/puppy-care-guide-for-new-parents.jpg', cv2.IMREAD_GRAYSCALE)

if grayscale_image is None:
    print("Error: Image file not found or path is incorrect.")
else:
    # Proceed with the rest of the code
    plt.imshow(grayscale_image, cmap='gray')
    plt.axis('off')
    plt.show()
    print(grayscale_image.dtype)
```
